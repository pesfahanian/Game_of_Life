```
import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
size = 100
pop_size = 10
iterations = 100
```

```
grid = np.array([[0 for j in range(size)] for i in range(size)])
plt.imshow(grid)
```

```
populated_grid = grid.copy()
populated_grid[50][30:70] = 1
plt.imshow(populated_grid)
```

```
new_grid = populated_grid.copy()

for itr in range(iterations):
    result_grid = new_grid.copy()
    for i in range(size):
        for j in range(size):

            # print("Point", i, ',', j)

            neighbourhood = [0, 0, 0, 0, 0, 0, 0, 0] # [Up, Down, Right, Left, Up_Right, Up_left, Down_Right, Down_Left]

            up_i = i-1
            up_j = j
            if(0 <= up_i <= size-1) and (0 <= up_j <= size-1):
                neighbourhood[0] = new_grid[up_i][up_j]

            down_i = i+1
            down_j = j
            if(0 <= down_i <= size-1) and (0 <= down_j <= size-1):
                neighbourhood[1] = new_grid[down_i][down_j]

            right_i = i
            right_j = j+1
            if(0 <= right_i <= size-1) and (0 <= right_j <= size-1):
                neighbourhood[2] = new_grid[right_i][right_j]

            left_i = i
            left_j = j-1
            if(0 <= left_i <= size-1) and (0 <= left_j <= size-1):
                neighbourhood[3] = new_grid[left_i][left_j]

            upright_i = i-1
            upright_j = j+1
            if(0 <= upright_i <= size-1) and (0 <= upright_j <= size-1):
                neighbourhood[4] = new_grid[upright_i][upright_j]

            upleft_i = i-1
            upleft_j = j-1
            if(0 <= upleft_i <= size-1) and (0 <= upleft_j <= size-1):
                neighbourhood[5] = new_grid[upleft_i][upleft_j]

            downright_i = i+1
            downright_j = j+1
            if(0 <= downright_i <= size-1) and (0 <= downright_j <= size-1):
                neighbourhood[6] = new_grid[downright_i][downright_j]

            downleft_i = i+1
            downleft_j = j-1
            if(0 <= downleft_i <= size-1) and (0 <= downleft_j <= size-1):
                neighbourhood[7] = new_grid[downleft_i][downleft_j]

            # print(neighbourhood)

            if(new_grid[i][j] == 0):
                if(sum(neighbourhood) >= 3):
                    # print('Happy')
                    result_grid[i][j] = 1
                else:
                    result_grid[i][j] = 0

            if(new_grid[i][j] == 1):
                if(sum(neighbourhood) <= 1):
                    # print('Lonely')
                    result_grid[i][j] = 0
                if(sum(neighbourhood) >= 4):
                    # print('Crowded')
                    result_grid[i][j] = 0
                if(2 <= sum(neighbourhood) <= 3):
                    # print('Same')
                    result_grid[i][j] = 1
    new_grid = result_grid.copy()

    plt.imshow(result_grid)
    plt.show()
```
